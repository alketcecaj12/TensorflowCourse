```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import numpy as np
```

```python
train_datagen = ImageDataGenerator(rescale = 1/255)
```

```python
train_generator = train_datagen.flow_from_directory("dati/labrador-bordercollie", 
                                  target_size = (300, 300), 
                                  batch_size = 64, 
                                  class_mode="binary")
```

```python
model = tf.keras.models.Sequential()

model.add(tf.keras.layers.Conv2D(16, (3, 3), activation = "relu", input_shape = (300, 300, 3)))
model.add(tf.keras.layers.MaxPooling2D((2, 2)))

model.add(tf.keras.layers.Conv2D(32, (3, 3), activation = "relu"))
model.add(tf.keras.layers.MaxPool2D((2, 2)))

model.add(tf.keras.layers.Conv2D(64, (3, 3), activation = "relu"))
model.add(tf.keras.layers.MaxPool2D((2, 2)))

model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(64, activation = "relu"))
model.add(tf.keras.layers.Dense(2))
```

```python
model.compile(optimizer = "adam", loss = tf.losses.SparseCategoricalCrossentropy(from_logits = True), 
             metrics = ["accuracy"])
```

```python
history = model.fit(train_generator, epochs = 30, validation_data = (train_generator))
```

```python
plt.figure(figsize = (20, 5))
plt.plot(history.history["accuracy"], label = "accuracy")
plt.plot(history.history["val_accuracy"], label = "val_accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend(loc = "lower right")
plt.show()
```

```python
plt.figure(figsize = (20, 5))
plt.plot(history.history["loss"], label = "loss")
plt.plot(history.history["val_loss"], label = "val_loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend(loc = "lower right")
plt.show()
```

```python
import io
from PIL import Image
import os
import numpy as np
from tensorflow.keras.preprocessing import image
```

```python
folder = "dati/files"

files = os.listdir(folder)

for f in files:
   
    if str(f) == ".DS_Store" : continue
    print(f)    
    img = image.load_img('dati/files/'+f, target_size=(300, 300))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    plt.imshow(img)
    plt.show()

    images = np.vstack([x])
    classes = model.predict(images, batch_size=10)

    if classes[0][0]<0.5:
        print("labrador")
    else:
        print("border collie")
       
    print('---------------------------------------') 
```
