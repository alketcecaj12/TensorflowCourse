```python
import numpy as np
import tensorflow as tf 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bmi =    [25, 30,  33,  24, 25,  25, 32,  31, 26, 25]
g =      [80, 120, 109, 83, 90, 100, 101, 102, 81, 79]
e =      [29, 59,  61 , 32, 31,  19, 66,  61, 26, 27]
esito =  [0,   1,   1,   0,  0,   0,  1,   1,  0,  0]
```

```python
data = pd.DataFrame({'bmi': bmi, 'g' : g, 'e': e, 'esito':esito})
```

```python
data
```

```python
X, y = data[['bmi', 'g', 'esito']].astype(float), data.iloc[:,-1]
```

```python
# create model
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(50, input_dim=3, activation='relu'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
# Compile model

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X, y, epochs = 100, batch_size = 64)
```

```python
plt.figure(figsize = (20, 4))
plt.plot(history.history['loss'])
plt.plot(history.history['accuracy'])
plt.xlabel('Number of Epochs')
plt.ylabel('Accuracy')
plt.legend([ 'loss', 'accuracy'], loc = 'upper right')

plt.show()
```
