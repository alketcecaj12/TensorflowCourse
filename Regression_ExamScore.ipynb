```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

### Define the problem
- a student has been sleeping and studying a certain number of hours. 
- how is his performance at the exam affected by this?

```python
sleep_study = np.array([[8,3],
                       [7,4],
                       [4,1],
                       [5,1],
                       [6,4],
                       [7,1]])
score = np.array([100, 108, 71, 73, 88, 91])
```

```python
sleep_study.shape
```

```python
model = tf.keras.Sequential([tf.keras.layers.Dense(units = 1, input_shape= [2])])
```

```python
model.compile(optimizer = 'sgd', loss = 'mean_squared_error')
```

```python
model.fit(sleep_study, score, epochs= 500)
```

```python
history = model.fit(sleep_study, score, epochs= 3000)
```

```python
plt.figure(figsize = (20, 5))
plt.plot(history.history['loss'])
plt.show()
```

```python
model.predict([[9, 4]])
```

```python
model.predict([[5, 1]])
```
