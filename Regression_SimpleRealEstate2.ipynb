```python
import numpy as np
import tensorflow as tf 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cam   = [     1,      2,      4,      2,      3,     3,     1,     5] # cam = 100.000
zona  = [     1,      1,      0,      1,      0,     1,     1,     0] # zona = {0, 1} + 15.000
parch = [     1,      0,      1,      1,      0,     0,     1,     0] # parch = {0,1} + 10.000
price = [125000, 215000, 410000, 225000, 300000,315000,125000,500000]
```

```python
dati = pd.DataFrame({'cam':cam, 'zona':zona, 'parch':parch, 'price':price})
```

```python
dati
```

```python
X = dati.iloc[:,: -1].values
y = dati.iloc[:, -1].values
```

```python
y.shape
```

```python
#model = tf.keras.Sequential([tf.keras.layers.Dense(units = 1, input_shape= [3])])
#model.compile(optimizer = 'sgd', loss = 'mean_squared_error')

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(1, input_shape = (8, 3), activation="relu"))


model.compile(optimizer = 'sgd', loss = 'mean_squared_error')
```

```python
history = model.fit(X, y, epochs = 500, verbose = 1)
```

```python
plt.figure(figsize = (20, 5))
plt.plot(history.history['loss'])
plt.show()
```

```python
model.predict([[4, 1, 0]])
```
