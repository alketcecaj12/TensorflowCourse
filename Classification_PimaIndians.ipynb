```python
import numpy as np
import tensorflow as tf 
import pandas as pd
import matplotlib.pyplot as plt

from tensorflow.keras.utils import plot_model
```

```python
data = pd.read_csv('dati/diabetes.csv')
```

```python
data
```

```python
data.corr()
```

```python
X, y = data[['Glucose', 'BMI', 'Age']].astype(float), data.iloc[:,-1]
```

```python
y
```

```python
y.shape
```

```python
# create model
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(1500, input_dim=3, activation='relu'))
model.add(tf.keras.layers.Dense(1000, input_dim=3, activation='relu'))
model.add(tf.keras.layers.Dense(500, input_dim=2, activation='relu'))
model.add(tf.keras.layers.Dense(100, input_dim=1, activation='relu'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
# Compile model

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X, y, epochs = 500, batch_size = 64)
```

```python
history.history
```

```python
plt.figure(figsize = (20, 4))
plt.plot(history.history['loss'])
plt.plot(history.history['accuracy'])
plt.xlabel('Number of Epochs')
plt.ylabel('Accuracy')
plt.legend([ 'loss', 'accuracy'], loc = 'upper right')

plt.show()
```
