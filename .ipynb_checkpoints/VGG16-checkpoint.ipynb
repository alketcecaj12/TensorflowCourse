```python
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load Keras VGG16 model that was pretrained against the ImageNet database 
model = VGG16()
from tensorflow.keras.applications import vgg16
```

```python
# load the image file by resizing it 224*224 as required by the model
img_path = 'dati/squalo.jpg'
```

```python
img = image.load_img(img_path, target_size=(224, 224))
plt.imshow(img)
```

```python
# convert the image to a numpy array
x = image.img_to_array(img)

# add a fourth dimension
x = np.expand_dims(x, axis=0)

# normalize the input images, 
x = preprocess_input(x)
```

```python
predictions = model.predict(x)
```

```python
pred_classes = vgg16.decode_predictions(predictions)
```

```python
for image_id, name, likelihood in pred_classes[0]:
    print('Predictions: {} - {:1f}'.format(name, likelihood))
```
