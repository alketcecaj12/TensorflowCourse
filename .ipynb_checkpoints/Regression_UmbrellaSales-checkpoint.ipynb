```python
import numpy as np
import tensorflow as tf 
import pandas as pd
import matplotlib.pyplot as plt

from tensorflow.keras.utils import plot_model
```

```python
data = pd.read_csv('dati/vendite_ombrelli.csv')
data
```

```python
plt.scatter(data['pioggia_mm'], data['ombrelli'])
plt.show()
```

```python
mm_pioggia = data['pioggia_mm'].values
nr_omb = data['ombrelli'].values
```

```python
X_train, X_test = mm_pioggia[0:400], mm_pioggia[400:]
y_train, y_test = nr_omb[0:400], nr_omb[400:]
```

```python
# Instantiate a Sequential model
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(500, input_shape=(1,),activation='relu'))
model.add(tf.keras.layers.Dense(100, input_shape=(1,),activation='relu'))
model.add(tf.keras.layers.Dense(50, input_shape=(1,),activation='relu'))
model.add(tf.keras.layers.Dense(1))
```

```python
# Compile your model
model.compile(optimizer = 'adam', loss = 'mse')

# Fit your model on your data for 30 epochs
history = model.fit(X_train, y_train, epochs = 500)
```

```python
plt.figure(figsize = (20, 5))
plt.plot(history.history['loss'])
plt.xlabel('Epoch')
plt.ylabel('Loss function')
plt.show()
```

```python
model.predict([37.3])
```

```python
previsioni = []
for i in X_test: 
    yhat = model.predict([i])
    print(yhat[0][0])
    previsioni.append(yhat[0][0])
```

```python
previsioni = np.array(previsioni)
```

```python
err = abs(previsioni - y_test)
```

```python
err_mean = np.mean(err)
```
